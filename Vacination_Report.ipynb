```python
# After Data will be cleaned and stored in a SQL databse
import pandas as pd
from sqlalchemy import create_engine

# Database connection (MySQL example)
# Format: mysql+pymysql://username:password@host:port/database
engine = create_engine("mysql+pymysql://root:vishwapriya@localhost:3306/vaccination_db")

# Load Excel files
coverage = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/coverage-data-cleaned.xlsx")
incidence = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/incidence-rate-data-cleaned.xlsx")
cases = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/reported-cases-data-cleaned.xlsx")
intro = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/vaccine-introduction-data-cleaned.xlsx")
schedule = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/vaccine-schedule-data-cleaned.xlsx")

# Store each dataset in SQL tables
coverage.to_sql("coverage_data", engine, if_exists="replace", index=False)
incidence.to_sql("incidence_rate_data", engine, if_exists="replace", index=False)
cases.to_sql("reported_cases_data", engine, if_exists="replace", index=False)
intro.to_sql("vaccine_introduction_data", engine, if_exists="replace", index=False)
schedule.to_sql("vaccine_schedule_data", engine, if_exists="replace", index=False)

print("✅ All datasets stored successfully in SQL database!")
```

```python
# Exploratory Data Analysis (EDA) for Vaccination Project

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load all datasets
coverage = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/coverage-data-cleaned.xlsx")
incidence = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/incidence-rate-data-cleaned.xlsx")
cases = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/reported-cases-data-cleaned.xlsx")
intro = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/vaccine-introduction-data-cleaned.xlsx")
schedule = pd.read_excel("C:/Users/VISHWA PRIYARANJAN/OneDrive/Desktop/Labmantrix_Internship/vaccine-schedule-data-cleaned.xlsx")

# ----------------------------
# 1. Statistical Summaries
# ----------------------------
print("==== Coverage Data Summary ====")
print(coverage.describe(include="all"))

print("\n==== Incidence Rate Summary ====")
print(incidence.describe(include="all"))

print("\n==== Reported Cases Summary ====")
print(cases.describe(include="all"))

# ----------------------------
# 2. Vaccination Coverage by Region
# ----------------------------
regional_coverage = coverage.groupby("GROUP")["COVERAGE"].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=regional_coverage, x="GROUP", y="COVERAGE", palette="viridis")
plt.title("Average Vaccination Coverage by Region")
plt.xlabel("Region")
plt.ylabel("Average Coverage (%)")
plt.xticks(rotation=45)
plt.show()

# ----------------------------
# 3. Disease Incidence Trends
# ----------------------------
trend = incidence.groupby("YEAR")["INCIDENCE_RATE"].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=trend, x="YEAR", y="INCIDENCE_RATE", marker="o")
plt.title("Global Disease Incidence Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Average Incidence Rate")
plt.grid(True)
plt.show()

# ----------------------------
# 4. Correlation Analysis
# ----------------------------
# Merge coverage & incidence data by Country + Year
merged = pd.merge(coverage, incidence,
                  left_on=["CODE","YEAR"], right_on=["CODE","YEAR"], how="inner")

corr = merged[["COVERAGE","INCIDENCE_RATE"]].corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Vaccination Coverage & Disease Incidence")
plt.show()

# ----------------------------
# 5. Regional Disparities
# ----------------------------
regional_disparities = merged.groupby("CODE")[["COVERAGE","INCIDENCE_RATE"]].mean().reset_index()

plt.figure(figsize=(12,6))
sns.scatterplot(data=regional_disparities, x="COVERAGE", y="INCIDENCE_RATE", hue="CODE")
plt.title("Regional Disparities: Coverage vs Disease Incidence")
plt.xlabel("Average Coverage (%)")
plt.ylabel("Average Incidence Rate")
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()

# ----------------------------
# 6. Insights
# ----------------------------
print("\n==== Insights from EDA ====")
print("1. Bar charts highlight which regions have lower vaccination coverage.")
print("2. Line graphs show global disease incidence trends declining over time.")
print("3. Heatmap shows negative correlation: higher coverage leads to lower incidence.")
print("4. Scatter plots reveal disparities between countries and regions.")
print("5. Helps identify regions where high incidence persists despite vaccine availability.")
```
